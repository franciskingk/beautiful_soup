```python
url = ["https://www.mandela.ac.za/", "https://www.unza.zm/" ]
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 6.3; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/54.0.2840.71 Safari/537.36'}
for item in url:
    r = requests.get(item, headers=headers)
    print(r.text)  
```

```python
#start page(s)
page = requests.get('https://www.unza.zm/') #'https://www.mandela.ac.za/'
page
```

```python
# printinf html content of the website
print(page.text)
```

```python
# parcing the html
soup = BeautifulSoup(page.text,"html.parser")
```

```python
#printing the content of the site in a pretier way
print(soup.prettify())
```

```python
results = soup.find_all('a')
results
```

```python
len(results)
#there are 87 links on this website
```

```python
# finding the text in the links
soup.find_all('a')[33].get_text()
```

```python
#checking results from the last item
results[-1]
```

```python
#
link_content = []
link_url = []

result = soup.find_all('a')

for result in results:
    #getting text out of every tag
    text= result.get_text()

    link = 'https://www.mandela.ac.za/' + result.get('href')

    # append to list
    link_content.append(text)

    # appending the text to the url link
    link_url.append(link)
```

```python
#preview
link_url[0:10]
```

```python
# saving the data
df=pd.DataFrame({"link_content":link_content,"link_url":link_url})
df.tail()
```
